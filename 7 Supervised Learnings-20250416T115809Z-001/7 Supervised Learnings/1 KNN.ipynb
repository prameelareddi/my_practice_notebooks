```python
# Consider the inpur varabile as X and Output variable as Y 
X = wbcd.iloc[:, 1:].values
y = wbcd.iloc[:, 0].values
X
```

```python
# Normalization
#X = (X-X.min())/(X.max() -X.min()) 
#X
```

 Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.80) 
```

 Feature Scaling (Very Important)

```python

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

Training the Model

```python
# import alogrithm method name from required libararies
from sklearn.neighbors import KNeighborsClassifier

# create an alogirthm ( same like as function)
classifier = KNeighborsClassifier(n_neighbors=3)  
```

```python
# apply the model on training dataset using fit
classifier.fit(X_train, y_train)
```

```python
# Predicting the Model : input variable of testing datset - xtest
y_pred = classifier.predict(X_test)
y_pred
```

```python
##Evaluating the Algorithm
from sklearn.metrics import confusion_matrix  

# creating confustion matrix table for TP and TN scenarios

print(confusion_matrix(y_test, y_pred))  
```

```python
from sklearn.metrics import accuracy_score 

# calculate the accuracy for the model by validating y pred and Y test

Accuracy_Score = accuracy_score(y_test, y_pred)
```
