```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris = pd.read_csv("D:\\Course\\Python\\Datasets\\iris.csv")
iris
```

```python
iris_setosa = iris.iloc[0:50,0:5] # 50
iris_versicolor = iris.iloc[50:100,0:5] # 50
iris_vergenica = iris.iloc[100:150,0:5] # 50 
```

```python
iris_versicolor
```

```python
# first 25 observation from setosa data - df1

df1 = pd.DataFrame(iris_setosa.iloc[0:25,0:5])
```

```python
# first 25 obesrvation from Versicolor data - df2

df2 = pd.DataFrame(iris_versicolor.iloc[0:25,0:5])
#df2
```

```python
# first 25 observations from Vergenica dataset - df3 

df3 = pd.DataFrame(iris_vergenica.iloc[0:25,0:5])
```

```python
# training dataset which contains 25 from all 3 datasets by using concat 

iris_train = pd.concat([df1,df2,df3])
```

```python
iris_train
```

```python
# Dividing the input and output variable for Training Dataset - xtrain and Y train
X_train = iris_train.iloc[:,0:4]
X_train
y_train = iris_train.iloc[:,4]
y_train
X_train
```

```python
# next 25 observation from Setosa dataset
df1 = pd.DataFrame(iris_setosa.iloc[25:50,0:5])
# next 25 observation from Versicolor dataset
df2 = pd.DataFrame(iris_versicolor.iloc[25:50,0:5])
# next 25 observation from Vergenica dataset
df3 = pd.DataFrame(iris_vergenica.iloc[25:50,0:5])
```

```python
# testing datset by concat the all 3 datasets ( df1,df2,df3)
iris_test= pd.concat([df1,df2,df3])
iris_test
```

```python
iris_test.shape
```

```python
X_test = iris_test.iloc[:,0:4]
X_test
y_test = iris_test.iloc[:,4]
y_test
```

```python
#  X_train and X_test are the dataset for training and Testing
#y_train ,y_test are target predictor for Training and Testing Dataset
```

```python
X_train, y_train,X_test,y_test
```

```python
# Fitting SVM to the Training set
# we dont have radial Method in python so we have to use linear method for linear Techniques
# Importing libraries and Method 
from sklearn.svm import SVC 
# Creating An alogrithm using imported method names
classifier = SVC(kernel = 'linear') 

#Apply alogrithm on Training dataset (xtrain , ytrain) by using .fitI()

classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results by applying on only input variables of 
# testind dataset (Xtest)

y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
# Making the Confusion Matrix by validating the predicted and Actual Values 

# importing required libraries 
from sklearn.metrics import confusion_matrix


cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score 

Accuracy_Score = accuracy_score(y_test, y_pred)
```

```python
Accuracy_Score
```
